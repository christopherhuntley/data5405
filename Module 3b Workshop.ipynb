````
# a dictionary-based word count function
def word_counts_dict(long_txt):
  words = long_txt.lower().replace("\n","").replace(".","").replace(",","").replace("—"," ").split()
  wc = {}
  for word in words:
    if word not in wc:
      wc[word] = 1
    else:
      wc[word] +=  1
  return wc

word_counts_dict(gettysburg)
````

## **Roman Numerals Yet Again.**
Work through exercise 2 from Lesson 9 (quoted below). Make sure everybody in your group understands both the rationale for the exercise and the solution. 

> Rewrite your `roman2int()` function from Lesson 6 to use a dictionary and an iterator instead of the 14-clause if statement. The dictionary should have 13 keys, one for each of the patterns.


````
#@title Click here for a solution

def roman2int(roman_numeral):
    rules = {"IV":4, "IX":9,"XL":40,"XC":90,"CD":400,"CM":900,
             "I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000}
    
    # initialize loop variables
    remaining = str(roman_numeral)
    int_value = 0
    
    while remaining: 
        # check all two-character rules
        if remaining[:2] in rules:
            int_value += rules[remaining[:2]]
            remaining = remaining[2:]
            
        # check all one-character rules    
        elif remaining[0] in rules:
            int_value += rules[remaining[0]]
            remaining = remaining[1:]
        # if no rule matches then consume the first character
        else:
            remaining = remaining[1:]
            
    return int_value

print(roman2int("XXIX"))
print(roman2int("MCMXCIX"))
print(roman2int("Super Bowl XXXIV"))
````

### Problem
Write a function `all_combinations()` that generates every subset (combination) of $r$ numbers from a sequence $0, ... , n-1$. For full credit, use a generator that yields each combination as a list. Otherwise, return a list of combinations as nested lists. 

### Notes
- There are ${n \choose r} = \frac{n!}{r!(n-r)!}$ possible combinations.
- The numbers within each subset $a, b, c, ...$ are always increasing with $a < b < c < ...$
- We can represent this as a matrix with ${n \choose r}$ rows (one per combination ) and $r$ columns (one per choice).
- The maximum number in each column $i$ is $n-r+i$. A column at the upper end of the range is _**maxed out**_.
- For example, with $n=7$ and $r=5$, the combinations are 
```python
[[0, 1, 2, 3, 4], # subsequences are always increasing
   [0, 1, 2, 3, 5],
   [0, 1, 2, 3, 6], # maximum of column 4 is 7-5+4 = 6
   [0, 1, 2, 4, 5],
   [0, 1, 2, 4, 6],
   [0, 1, 2, 5, 6], # maximum of column 3 is 7-5+3 = 5
   [0, 1, 3, 4, 5],
   [0, 1, 3, 4, 6],
   [0, 1, 3, 5, 6],
   [0, 1, 4, 5, 6], # maximum of column 2 is 7-5+2 = 4
   [0, 2, 3, 4, 5],
   [0, 2, 3, 4, 6],
   [0, 2, 3, 5, 6],
   [0, 2, 4, 5, 6],
   [0, 3, 4, 5, 6], # maximum of column 1 is 7-5+1 = 3
   [1, 2, 3, 4, 5],
   [1, 2, 3, 4, 6],
   [1, 2, 3, 5, 6],
   [1, 2, 4, 5, 6],
   [1, 3, 4, 5, 6],
   [2, 3, 4, 5, 6]]  # maximum of column 0 is 7-5+0 = 2
 ```

### General Algorithm

Treat this as a constrained counting problem, with each matrix row representing a number with $k$ digits. Starting in the first row with each column at its lowest possible value (i.e., [0,1, ..., $r$-2, $r$-1]) we iterate through the possible combinations one at a time, stopping in the last row with every column maxed out (i.e., [$n-r$, $n-r+1$, ..., $n$-2, $n$-1]). In each iteration, we _**increment**_ the numbers in the preceeding row as follows:
- Select the right-most column that is not maxed out. Mark it as the _**increment column**_.
- Add 1 to the selected _increment column_.
- All columns to the _right_ of the _increment column_ are _**reset**_ to count upwards from the current value of the increment column. If the number in the increment column is 3, then the numbers in the columns to its right are 4, 5, 6 etc. 

We stop when column 0 is maxed out. 

So, let's step through the ($n$=7, $r$=5) example, one iteration (row) at a time:
```python
   [0, 1, 2, 3, 4]  # initial sequence 
   [0, 1, 2, 3, 5]  # incr col 4
   [0, 1, 2, 3, 6]  # incr col 4; col 4 is maxed out
   [0, 1, 2, 4, 5]  # incr col 3, reset col 4 to 4+1 = 5
   [0, 1, 2, 4, 6]  # incr col 4; col 4 is maxed out
   [0, 1, 2, 5, 6]  # incr col 3; reset col 4 to 5+1 = 6; cols 3 and 4 are maxed out
   [0, 1, 3, 4, 5]  # incr col 2; reset cols 3 and 4
   [0, 1, 3, 4, 6]  # incr col 4; col 4 is maxed out
   [0, 1, 3, 5, 6]  # incr col 3; reset col 4; cols 3 and 4 are maxed out
   [0, 1, 4, 5, 6]  # incr col 2; reset cols 3 & 4; cols 2-4 are maxed out
   [0, 2, 3, 4, 5]  # incr col 1; reset cols 2-4
   [0, 2, 3, 4, 6]  # incr col 4; col 4 is maxed out
   [0, 2, 3, 5, 6]  # incr col 3; cols 3 and 4 are maxed out
   [0, 2, 4, 5, 6]  # incr col 2; cols 2-4 are maxed out
   [0, 3, 4, 5, 6]  # incr col 1; reset cols 2-4; cols 1-4 are maxed out
   [1, 2, 3, 4, 5]  # incr col 0; reset cols 1-4
   [1, 2, 3, 4, 6]  # incr col 4; col 4 is maxed out
   [1, 2, 3, 5, 6]  # incr col 3; reset col 4; cols 3 and 4 are maxed out
   [1, 2, 4, 5, 6]  # incr col 2; reset cols 3 and 4; cols 2-4 maxed out
   [1, 3, 4, 5, 6]  # incr col 1; reset cols 2-4; cols 1-4 maxed out
   [2, 3, 4, 5, 6]  # incr col 0; reset cols 1-4; all cols maxed out
 ```
